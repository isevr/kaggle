```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
df = pd.read_csv("vgsales.csv")
df
```

```python
df.Year.fillna(2007, inplace=True)
df.Publisher.fillna("Nintendo", inplace=True)
```

```python
plat = pd.DataFrame.to_numpy(pd.get_dummies(df.Platform))
genre = pd.DataFrame.to_numpy(pd.get_dummies(df.Genre))
pub = pd.DataFrame.to_numpy(pd.get_dummies(df.Publisher.unique()))
```

```python
clf =RandomForestClassifier(random_state=0)
clf.fit([[plat],[genre]],[pub])
```
